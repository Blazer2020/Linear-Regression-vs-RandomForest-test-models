# **Test ML project**

# **Load data**


```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv')
df
```

# **Data preparation**


## Data Seperation as X and Y

```python
y = df['logS']
y
```

```python
x = df.drop('logS',axis=1) #axis =1 mean column mode axis=0 is row mode
x
```

## **Data Splitting**

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=100) #20% would be test data random_state=100 mean way we split would be fixed on aanother time we split also same data will be test data and same data will be train data
```

```python
x_train
```

```python
x_test
```

#**Model Building**

## **Linear Regression**

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train,y_train)
```

### **Applying Model to make Prediction**

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

###**Evaluate performance**

```python
from sklearn.metrics import mean_squared_error, r2_score #squared correlation coefficient

lr_train_mse = mean_squared_error(y_train,y_lr_train_pred)
lr_train_r2 = r2_score(y_train,y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test,y_lr_test_pred)
lr_test_r2 = r2_score(y_test,y_lr_test_pred)
```

```python
print('LR MSE (Train):',lr_train_mse)
print('LR R2 (Train):',lr_train_r2)
print('LR MSE (Test):',lr_test_mse)
print('LR R2 (Test):',lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear Regression',lr_train_mse,lr_train_r2,lr_test_mse,lr_test_r2]).transpose()
lr_results.columns = ['Method','Training MSE','Training R2','Test MSE','Test R2']
lr_results
```

```python
new_input = [[1.5, 300, 5,0.5]]
predicted_price = lr.predict(new_input)
print("Predicted price:", predicted_price[0])
```

##**Random Forest**

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train,y_train)
```

###**Applying the model**

```python
y_rf_train_pred = lr.predict(x_train)
y_rf_test_pred = lr.predict(x_test)
```

###**Evaluate Performance**

```python
from sklearn.metrics import mean_squared_error, r2_score #squared correlation coefficient

rf_train_mse = mean_squared_error(y_train,y_rf_train_pred)
rf_train_r2 = r2_score(y_train,y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test,y_rf_test_pred)
rf_test_r2 = r2_score(y_test,y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random Forest',rf_train_mse,rf_train_r2,rf_test_mse,rf_test_r2]).transpose()
rf_results.columns = ['Method','Training MSE','Training R2','Test MSE','Test R2']
rf_results
```

#**Model Comparison**

```python
df_models = pd.concat([lr_results,rf_results],axis=0)
df_models.reset_index(drop=True,inplace=True)
df_models
```
